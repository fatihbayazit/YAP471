```python
#BOLUM2

# DosyayÄ± oku
df = pd.read_csv("stratejiler2.csv")

# Her bir nokta Ã§ifti iÃ§in kontrol yap
for i in range(len(df)):
    for j in range(i + 1, len(df)):
        x1, y1, price1 = df.loc[i, ["X", "Y", "Hissenin Spot Fiyati"]]
        x2, y2, price2 = df.loc[j, ["X", "Y", "Hissenin Spot Fiyati"]]
        
        # Ä°ki nokta arasÄ±ndaki mesafeyi hesapla
        distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
        
        # Hisse fiyatlarÄ±nÄ±n arasÄ±ndaki farkÄ± kontrol et
        price_difference = abs(price2 - price1)
        
        # EÄer mesafe hisse fiyatlarÄ±nÄ±n farkÄ±nÄ± aÅÄ±yorsa strateji mÃ¼mkÃ¼n
        if distance > price_difference:
            print(f"Strateji mÃ¼mkÃ¼n: Nokta {i+1} ve Nokta {j+1}")
        else:
            print(f"Strateji imkansÄ±z: Nokta {i+1} ve Nokta {j+1}")
```

```python
#BOLUM3

#Soru 1
#Bu soru iÃ§in uygun bir strateji, iron condor stratejisidir. Bu strateji, hisse senedi fiyatÄ±nÄ±n belirli bir aralÄ±kta kalacaÄÄ±nÄ± dÃ¼ÅÃ¼nen yatÄ±rÄ±mcÄ±lar iÃ§in uygundur. 
#Strike fiyatÄ± A TL olan bir put opsiyonu sat.
#Strike fiyatÄ± B TL olan bir put opsiyonu al. (B < A)
#Strike fiyatÄ± C TL olan bir call opsiyonu sat.
#Strike fiyatÄ± D TL olan bir call opsiyonu al. (C < D)
#Bu strateji, hisse senedi fiyatÄ± B TL ile C TL arasÄ±nda olduÄunda, satÄ±lan put ve call opsiyonlarÄ±nÄ±n primlerinin toplamÄ±nÄ± kar olarak elde etmenizi saÄlar. Hisse senedi fiyatÄ± B TLânin altÄ±na veya C TLânin Ã¼zerine Ã§Ä±ktÄ±ÄÄ±nda, alÄ±nan put ve call opsiyonlarÄ±nÄ±n primleri, satÄ±lan opsiyonlarÄ±n zararÄ±nÄ± sÄ±nÄ±rlamaya baÅlar. Hisse senedi fiyatÄ± A TLânin altÄ±na veya D TLânin Ã¼zerine Ã§Ä±ktÄ±ÄÄ±nda, stratejinin zararÄ± maksimuma ulaÅÄ±r ve A - B - primlerin toplamÄ± veya D - C - primlerin toplamÄ± olur. Hisse senedi fiyatÄ± A TL ile B TL arasÄ±nda veya C TL ile D TL arasÄ±nda olduÄunda, stratejinin kar/zararÄ± deÄiÅir.

A = 70 # satÄ±lan put opsiyonunun strike fiyatÄ± k2
B = 80 # alÄ±nan put opsiyonunun strike fiyatÄ± k1
C = 120 # satÄ±lan call opsiyonunun strike fiyatÄ± k3
D = 130 # alÄ±nan call opsiyonunun strike fiyatÄ± k4
P = 3 # satÄ±lan put opsiyonunun primi
Q = 5 # alÄ±nan put opsiyonunun primi
R = 2 # satÄ±lan call opsiyonunun primi
S = 4 # alÄ±nan call opsiyonunun primi
x = [0, A, B, C, D, 200] # hisse senedi fiyatlarÄ±
premium = P + R - Q - S
y = [premium,premium, B - A + premium,B - A + premium, B - A - (D - C) + premium,B - A - (D - C) + premium] # strateji kar/zararlarÄ±

# grafiÄi Ã§iz
plt.plot(x, y, color="orange", marker="o")
plt.xlabel("Stock Price (TL)")
plt.ylabel("Profit/Loss (TL)")
plt.title("Iron Condor Strategy")
plt.show()            

#Soru 2
#Bu soru iÃ§in uygun bir strateji, bull call spread stratejisidir. Bu strateji, hisse senedi fiyatÄ±nÄ±n hafif bir yÃ¼kseliÅ gÃ¶stereceÄini bekleyen yatÄ±rÄ±mcÄ±lar iÃ§in uygundur.
#Strike fiyatÄ± E TL olan bir call opsiyonu al.
#Strike fiyatÄ± F TL olan bir call opsiyonu sat. (E < F)
#Bu strateji, hisse senedi fiyatÄ± E TLânin Ã¼zerine Ã§Ä±ktÄ±ÄÄ±nda, alÄ±nan call opsiyonunun primi, satÄ±lan call opsiyonunun zararÄ±nÄ± karÅÄ±lamaya baÅlar. Hisse senedi fiyatÄ± F TLâye ulaÅtÄ±ÄÄ±nda, stratejinin karÄ± maksimuma ulaÅÄ±r ve F - E - primlerin farkÄ± olur. Hisse senedi fiyatÄ± F TLânin Ã¼zerine Ã§Ä±ktÄ±ÄÄ±nda, stratejinin karÄ± sabit kalÄ±r. Hisse senedi fiyatÄ± E TLânin altÄ±nda olduÄunda, stratejinin zararÄ± maksimuma ulaÅÄ±r ve primlerin farkÄ± olur.

# strateji verilerini tanÄ±mla
E = 90 # alÄ±nan call opsiyonunun strike fiyatÄ±
F = 100 # satÄ±lan call opsiyonunun strike fiyatÄ±
T = 6 # alÄ±nan call opsiyonunun primi
U = 3 # satÄ±lan call opsiyonunun primi
x = [0, E, F, 200] # hisse senedi fiyatlarÄ±
y = [-T + U, -T + U, F - E - T + U, F - E - T + U] # strateji kar/zararlarÄ±

# grafiÄi Ã§iz
plt.plot(x, y, color="blue", marker="o")
plt.xlabel("Stock Price (TL)")
plt.ylabel("Profit/Loss (TL)")
plt.title("Bull Call Spread Strategy")
plt.show()

#Soru 3
#Bu soru iÃ§in uygun bir strateji, long put stratejisidir. Bu strateji, hisse senedi fiyatÄ±nÄ±n ciddi bir dÃ¼ÅÃ¼Å gÃ¶stereceÄini bekleyen yatÄ±rÄ±mcÄ±lar iÃ§in uygundur.
#Strike fiyatÄ± G TL olan bir put opsiyonu al.
#Bu strateji, hisse senedi fiyatÄ± G TLânin altÄ±na indiÄinde, alÄ±nan put opsiyonunun primi, hisse senedinin zararÄ±nÄ± karÅÄ±lamaya baÅlar. Hisse senedi fiyatÄ± dÃ¼ÅtÃ¼kÃ§e, stratejinin karÄ± artar. Hisse senedi fiyatÄ± G TLânin Ã¼zerinde olduÄunda, stratejinin zararÄ± maksimuma ulaÅÄ±r ve prim olur.

# strateji verilerini tanÄ±mla
G = 110 # alÄ±nan put opsiyonunun strike fiyatÄ±
V = 8 # alÄ±nan put opsiyonunun primi
x = [0, G, 200] # hisse senedi fiyatlarÄ±
y = [G - V, -V, -V] # strateji kar/zararlarÄ±

# grafiÄi Ã§iz
plt.plot(x, y, color="red", marker="o")
plt.xlabel("Stock Price (TL)")
plt.ylabel("Profit/Loss (TL)")
plt.title("Long Put Strategy")
plt.show()
```
